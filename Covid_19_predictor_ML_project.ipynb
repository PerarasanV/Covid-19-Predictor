Recursive Feature Elimination and Cross validation

RFECV TO FIND OUT THE BEST NUMBER OF FEATURES REQUIRED FOR PREDICTION/MODEL TRAINING.

```python
from sklearn.feature_selection import RFECV
from sklearn.model_selection import StratifiedKFold
```

```python
model=LogisticRegression()          #USING LOGISTIC REGRESSION FOR RFECV
```

```python
X=df_4.drop('Corona',axis=1)
y=df_4[['Corona']]
```

```python
rfecv=RFECV(estimator=model,step=1,cv=StratifiedKFold(5),scoring='accuracy')
```

```python
rfecv.fit(X,y)
```

```python
optimal_features = [f"Feature {i+1}" for i, support in enumerate(rfecv.support_) if support]
```

```python
print("Optimal number of features: {}".format(rfecv.n_features_))
print("Optimal features: {}".format(optimal_features))
```

```python
#RFECV concludes that there are 3 optimal feature for model training.
```

```python
#SMOTE will be used since there is a slight imbalance in the target class.this will balance out the imbalance in target variable.
```

```python
X_2=df_4[['Cough_symptoms','Sore_throat','Shortness_of_breath','Sex']]  #optimal features for model training.
y_2=df_4[['Corona']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_2, y_2,stratify=y_2,test_size=0.2, random_state=42)
```

```python
from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE(sampling_strategy='auto', random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_train, y_train)
```

MULTIPLE MODELS WILL BE TESTED TO CHOOSE THE BEST ONE.

```python
## Model 1 ---- LOGISTIC REGRESSION.
```

```python
model_1= LogisticRegression(C=1.0, random_state=42)
```

```python
model_1.fit(X_resampled,y_resampled)
```

```python
y_pred=model_1.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy=accuracy_score(y_test,y_pred)
```

```python
accuracy
```

```python
print(f"Accuracy: {accuracy:.4f}")
```
